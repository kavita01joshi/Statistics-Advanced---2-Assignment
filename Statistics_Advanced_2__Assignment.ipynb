###Q1- What is hypothesis testing in statistics?

* Hypothesis testing is a statistical method that evaluates assumptions (hypotheses) about a population parameter using sample data. It helps determine whether the observed data provides enough evidence to reject a stated assumption about the population

**Key Components**

Null Hypothesis (Hâ‚€): The default assumption that there is no effect or no difference. For example, "The average height of boys and girls at age 5 is the same."

Alternative Hypothesis (Hâ‚ or Hâ‚): The claim that contradicts the null hypothesis. For example, "Girls are shorter than boys at age 5."

Significance Level (Î±): The threshold for deciding whether to reject Hâ‚€, commonly set at 0.05 (5%).

p-value: The probability of observing the data if Hâ‚€ is true. A small p-value (less than Î±) suggests rejecting Hâ‚€.

Test Statistic: A calculated value from the sample data used to decide between Hâ‚€ and Hâ‚.

Critical Value: A cutoff point that helps determine whether the test statistic falls in the rejection region.

**Types of Hypothesis Tests**

One-Tailed Test: Tests for a difference in one direction (e.g., greater than or less than).

Two-Tailed Test: Tests for a difference in either direction (e.g., not equal to).

**Common Errors**

Type I Error (Î±): Rejecting Hâ‚€ when it's actually true.

Type II Error (Î²): Failing to reject Hâ‚€ when it's actually false.

###Q2. What is the null hypothesis, and how does it differ from the alternative hypothesis?

**Null Hypothesis (Hâ‚€)**

* The null hypothesis is a default assumption that there is no effect, no difference, or no relationship between variables.

* It represents the status quo or a baseline claim.

* Example: "The average test scores of students before and after tutoring are the same."

**Alternative Hypothesis (Hâ‚ or Hâ‚)**

* The alternative hypothesis is what you want to test or prove.

* It suggests that there is an effect, a difference, or a relationship.

* Example: "Students score higher after tutoring than before."


**Feature**	  >>   **Null Hypothesis (Hâ‚€)** 	>>  **Alternative Hypothesis (Hâ‚ or Hâ‚)**

Assumption	>> No change or effect	>> There is a change or effect

Goal of Testing >>	Try to reject it	>> Try to support it

Basis for Decision >>	Statistical evidence >>	Statistical evidence

Example >> Claim	"No difference in blood pressure"	>> "Blood pressure decreases with medication"

**Why It Matters**

In hypothesis testing, you never prove the alternative hypothesis directly. Instead, you gather enough evidence to reject the null hypothesis, which indirectly supports the alternative.

###Q3. Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.

**Significance Level**-

* The significance level (Î±) is the probability of making a Type I error â€” that is, rejecting the null hypothesis (Hâ‚€) when it is actually true.

* It sets the threshold for how much evidence you require to reject Hâ‚€.

Common values:

0.05 (5%) â†’ most widely used

0.01 (1%) â†’ stricter, used in high-stakes testing

0.10 (10%) â†’ more lenient, used in exploratory research

**How It Works in a Test**

Set Î± before testing â€” this is your tolerance for risk.

Calculate the p-value from your sample data.

**Compare p-value to Î±:**

If p-value â‰¤ Î± â†’ reject Hâ‚€ (evidence is strong enough)

If p-value > Î± â†’ fail to reject Hâ‚€ (not enough evidence)

**Why It Matters**

Controls false positives: A lower Î± reduces the chance of wrongly rejecting Hâ‚€.

Balances risk: Choosing Î± depends on how serious a Type I error would be in your context (e.g., medical trials vs. marketing tests).

Guides decision-making: Itâ€™s the benchmark for interpreting your results.

**Example**-

Suppose you're testing a new drug. You set Î± = 0.01 because a false positive could be dangerous. If your p-value is 0.008, you reject Hâ‚€ and conclude the drug has a significant effect.

###Q4.  What are Type I and Type II errors? Give examples of each.
**Type I Error (False Positive)**

* Definition: Rejecting the null hypothesis (Hâ‚€) when it is actually true.

* You think there's an effect or difference, but there isn't.

* Controlled by: The significance level (Î±) â€” usually set at 0.05.

ðŸ§ª Example:
Imagine a medical test for a new drug:

Hâ‚€: The drug has no effect.

You conduct a test and reject Hâ‚€, concluding the drug works.

But in reality, the drug doesn't work â€” you've made a Type I error.

**Type II Error (False Negative)**
* Definition: Failing to reject the null hypothesis (Hâ‚€) when it is actually false.

* You miss a real effect or difference.

* Controlled by: The power of the test (1 âˆ’ Î²), which depends on sample size, effect size, and Î±.

ðŸ§ª Example:
Same drug scenario:

Hâ‚€: The drug has no effect.

You conduct a test and fail to reject Hâ‚€, concluding the drug doesn't work.

But in reality, the drug does work â€” you've made a Type II error.

**Comparison**
**Error Type** >>	**What Happens** >>	**Reality** >>	**Consequence**

Type I Error >>	Reject Hâ‚€ when it's true	>> No real effect	>> False alarm

Type II Error >>	Fail to reject Hâ‚€ when it's false	>> Real effect exists	>>Missed opportunity

###Q5. What is the difference between a Z-test and a T-test? Explain when to use each.

**Use a Z-Test when:**

* You have a large sample size (n â‰¥ 30).

* The population standard deviation (Ïƒ) is known.

* You're testing population means or proportions.

Example: Testing whether the average height of adult males in a city is 170 cm, using a sample of 100 men and known population standard deviation.

**Use a T-Test when:**

* You have a small sample size (n < 30).

* The population standard deviation is unknown.

* You're comparing sample means (one-sample, two-sample, or paired).

Example: Comparing the test scores of two small groups of students to see if a new teaching method is effective.


**Z-Test vs T-Test: Key Differences**

Feature 	>>  Z-Test  >> 	T-Test

Sample Size	>> Large (typically n â‰¥ 30)	>> Small (typically n < 30)

Population >> Std Dev	 >> Known	Unknown

Distribution Used	>> Standard Normal (Z) >> Distribution	Studentâ€™s T Distribution

Precision	 >> More precise with large samples	>> Adjusts for uncertainty in small samples
