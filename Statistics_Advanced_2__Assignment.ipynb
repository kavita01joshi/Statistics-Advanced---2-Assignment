###Q1- What is hypothesis testing in statistics?

* Hypothesis testing is a statistical method that evaluates assumptions (hypotheses) about a population parameter using sample data. It helps determine whether the observed data provides enough evidence to reject a stated assumption about the population

**Key Components**

Null Hypothesis (H₀): The default assumption that there is no effect or no difference. For example, "The average height of boys and girls at age 5 is the same."

Alternative Hypothesis (H₁ or Hₐ): The claim that contradicts the null hypothesis. For example, "Girls are shorter than boys at age 5."

Significance Level (α): The threshold for deciding whether to reject H₀, commonly set at 0.05 (5%).

p-value: The probability of observing the data if H₀ is true. A small p-value (less than α) suggests rejecting H₀.

Test Statistic: A calculated value from the sample data used to decide between H₀ and H₁.

Critical Value: A cutoff point that helps determine whether the test statistic falls in the rejection region.

**Types of Hypothesis Tests**

One-Tailed Test: Tests for a difference in one direction (e.g., greater than or less than).

Two-Tailed Test: Tests for a difference in either direction (e.g., not equal to).

**Common Errors**

Type I Error (α): Rejecting H₀ when it's actually true.

Type II Error (β): Failing to reject H₀ when it's actually false.

###Q2. What is the null hypothesis, and how does it differ from the alternative hypothesis?

**Null Hypothesis (H₀)**

* The null hypothesis is a default assumption that there is no effect, no difference, or no relationship between variables.

* It represents the status quo or a baseline claim.

* Example: "The average test scores of students before and after tutoring are the same."

**Alternative Hypothesis (H₁ or Hₐ)**

* The alternative hypothesis is what you want to test or prove.

* It suggests that there is an effect, a difference, or a relationship.

* Example: "Students score higher after tutoring than before."


**Feature**	  >>   **Null Hypothesis (H₀)** 	>>  **Alternative Hypothesis (H₁ or Hₐ)**

Assumption	>> No change or effect	>> There is a change or effect

Goal of Testing >>	Try to reject it	>> Try to support it

Basis for Decision >>	Statistical evidence >>	Statistical evidence

Example >> Claim	"No difference in blood pressure"	>> "Blood pressure decreases with medication"

**Why It Matters**

In hypothesis testing, you never prove the alternative hypothesis directly. Instead, you gather enough evidence to reject the null hypothesis, which indirectly supports the alternative.

###Q3. Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.

**Significance Level**-

* The significance level (α) is the probability of making a Type I error — that is, rejecting the null hypothesis (H₀) when it is actually true.

* It sets the threshold for how much evidence you require to reject H₀.

Common values:

0.05 (5%) → most widely used

0.01 (1%) → stricter, used in high-stakes testing

0.10 (10%) → more lenient, used in exploratory research

**How It Works in a Test**

Set α before testing — this is your tolerance for risk.

Calculate the p-value from your sample data.

**Compare p-value to α:**

If p-value ≤ α → reject H₀ (evidence is strong enough)

If p-value > α → fail to reject H₀ (not enough evidence)

**Why It Matters**

Controls false positives: A lower α reduces the chance of wrongly rejecting H₀.

Balances risk: Choosing α depends on how serious a Type I error would be in your context (e.g., medical trials vs. marketing tests).

Guides decision-making: It’s the benchmark for interpreting your results.

**Example**-

Suppose you're testing a new drug. You set α = 0.01 because a false positive could be dangerous. If your p-value is 0.008, you reject H₀ and conclude the drug has a significant effect.

###Q4.  What are Type I and Type II errors? Give examples of each.
**Type I Error (False Positive)**

* Definition: Rejecting the null hypothesis (H₀) when it is actually true.

* You think there's an effect or difference, but there isn't.

* Controlled by: The significance level (α) — usually set at 0.05.

🧪 Example:
Imagine a medical test for a new drug:

H₀: The drug has no effect.

You conduct a test and reject H₀, concluding the drug works.

But in reality, the drug doesn't work — you've made a Type I error.

**Type II Error (False Negative)**
* Definition: Failing to reject the null hypothesis (H₀) when it is actually false.

* You miss a real effect or difference.

* Controlled by: The power of the test (1 − β), which depends on sample size, effect size, and α.

🧪 Example:
Same drug scenario:

H₀: The drug has no effect.

You conduct a test and fail to reject H₀, concluding the drug doesn't work.

But in reality, the drug does work — you've made a Type II error.

**Comparison**
**Error Type** >>	**What Happens** >>	**Reality** >>	**Consequence**

Type I Error >>	Reject H₀ when it's true	>> No real effect	>> False alarm

Type II Error >>	Fail to reject H₀ when it's false	>> Real effect exists	>>Missed opportunity

###Q5. What is the difference between a Z-test and a T-test? Explain when to use each.

**Use a Z-Test when:**

* You have a large sample size (n ≥ 30).

* The population standard deviation (σ) is known.

* You're testing population means or proportions.

Example: Testing whether the average height of adult males in a city is 170 cm, using a sample of 100 men and known population standard deviation.

**Use a T-Test when:**

* You have a small sample size (n < 30).

* The population standard deviation is unknown.

* You're comparing sample means (one-sample, two-sample, or paired).

Example: Comparing the test scores of two small groups of students to see if a new teaching method is effective.


**Z-Test vs T-Test: Key Differences**

Feature 	>>  Z-Test  >> 	T-Test

Sample Size	>> Large (typically n ≥ 30)	>> Small (typically n < 30)

Population >> Std Dev	 >> Known	Unknown

Distribution Used	>> Standard Normal (Z) >> Distribution	Student’s T Distribution

Precision	 >> More precise with large samples	>> Adjusts for uncertainty in small samples
